Q1. Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.

Also mention any two industries where Tableau is commonly used and explain one use case.

Ans: Tableau is a powerful data visualization and business intelligence (BI) tool used to analyze data and create interactive dashboards and reports. It helps users convert raw data into visual formats such as charts, graphs, maps, and dashboards, making it easier to understand trends and patterns.

Tableau allows users to connect to different data sources like Excel files, SQL databases, cloud services, and big data platforms. It uses a drag-and-drop interface, so even non-programmers can perform data analysis easily.

Tableau is preferred over traditional spreadsheet tools

1.Better Data Visualization:
- Tableau can create advanced charts, interactive dashboards, and maps, while spreadsheet tools usually provide limited visualization options.

2.Handles Large Data Efficiently
- Tableau can analyze millions of rows of data without slowing down, whereas spreadsheets like Excel may become slow with large datasets.

3.Interactive Dashboards
- Users can filter, drill down, and interact with data in real time, which is difficult in traditional spreadsheets.

4.Easy Data Integration
- Tableau can connect to multiple data sources at once and combine them for analysis.

5.Faster Data Analysis
- Drag-and-drop features allow users to build visualizations quickly without writing complex formulas.

Two industries where Tableau is commonly used

Retail Industry

Healthcare Industry

Example Use Case (Retail Industry)

In the retail industry, Tableau is used to analyze sales performance and customer buying behavior. Retail companies can create dashboards showing:

Sales by product category
Sales by region or store
Monthly or yearly sales trends
Customer purchasing patterns

For example, a retail company can identify which product category generates the highest sales and profit and which region has the highest demand. This helps management make better decisions about inventory management, marketing strategies, and pricing.

1.Open Tableau and connect to the Sample – Superstore dataset.

2. Which field goes to Columns?

- Sales goes to the Columns shelf.

This means Tableau adds up all the sales values for each category to show the total sales per category.

Q8.What is the difference between a Worksheet and a Dashboard in Tableau?

Why is a dashboard more useful for decision-makers?

Ans:The difference between worksheet and dashboard are

| Feature |   worksheet | Dashboard |
| ------------- | ------------------------------------------------------------------------- | ------------------------------------------------------------------------------ |
| Definition    | A **Worksheet** is a single visualization such as a chart, graph, or map. | A **Dashboard** is a collection of multiple worksheets combined in one screen. |
| Purpose       | Used to create and analyze **one visualization at a time**.               | Used to **present multiple visualizations together** for better analysis.      |
| Content       | Contains **one chart or graph**.                                          | Contains **multiple charts, filters, images, and text**.                       |
| Interactivity | Limited interaction with one dataset.                                     | Highly interactive with filters and actions across multiple views.             |
| Use Case      | Used mainly for **data exploration and analysis**.                        | Used mainly for **presentation and decision making**.   |

Dashboard is More Useful for Decision-Makers

A dashboard is more useful because it shows multiple insights in one place, making it easier for managers and executives to understand business performance quickly.

Advantages for decision-makers:

1.Complete view of business data – Multiple charts show sales, profit, and trends together.

2.Quick decision making – Important information is visible in one screen.

3.Interactive analysis – Users can apply filters and explore specific data.

4.Better data storytelling – Dashboards present data in a clear and visual way.

Q9.What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.

Ans:Tableau Prep is a data preparation tool used to clean, combine, and organize raw data before analysis in Tableau. It helps users prepare data by fixing errors, removing unwanted values, and transforming datasets into a format that is easier to analyze.

Tableau Prep provides a visual and step-by-step workflow, allowing users to perform data cleaning tasks without complex coding.

 Tableau Prep is important in real-world data projects

In real-world projects, raw data is often messy and inconsistent. Tableau Prep helps analysts to:

1.Clean inaccurate or incomplete data

2.Combine data from multiple sources

3.Remove duplicates and errors

4.Standardize formats and data types

5.Prepare reliable data for dashboards and reports

This ensures that the final analysis and dashboards in Tableau are accurate and trustworthy for business decision-making.

Data Cleaning Tasks in Tableau Prep
1. Remove Null Values

Steps:

1.Open the dataset in Tableau Prep.

2.Add a Clean Step in the workflow.

3.Select the field containing null values.

4.In the profile pane, click on the Null values section.

5.Choose Filter → Exclude Null Values.

Result:
All records containing null values in that field will be removed.

2. Remove Duplicate Records

Steps:

1.Load the dataset in Tableau Prep.

2.Add a Clean Step.

3.Select the columns that should contain unique values (for example: Order ID).

4.Use Group and Replace → Remove Duplicate Rows or apply Aggregate to keep unique records.

Result:

Duplicate rows will be removed, leaving only one unique record for each entry.

3. Change Incorrect Data Types

Steps:

1.Open the dataset in Tableau Prep.

2.In the Profile Pane, locate the field with the wrong data type.

3.Click on the data type icon beside the field name.

4.Select the correct data type (for example: change Text → Number or Text → Date).

5.Verify that the data is correctly formatted.
